```python
#@title Exhibit Vivisted Counts

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Visitor_Logs.csv')

df['Entry Time'] = pd.to_datetime(df['Entry Time'])
df['Exit Time'] = pd.to_datetime(df['Exit Time'])
df['Visit Duration'] = df['Exit Time'] - df['Entry Time']

exhibit_counts = df['Exhibit Visited'].value_counts()
exhibit_counts.plot(kind='bar', xlabel='Exhibit', ylabel='Number of Visits', title='Exhibit Visited Counts')
plt.show()
```

```python
#@title Location Frequencies

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Security_Camera_Logs.csv')

df['Timestamp'] = pd.to_datetime(df['Timestamp'])

most_frequent_location = df['Location'].mode().values[0]
print(f"The most frequent location is {most_frequent_location}")

location_counts = df['Location'].value_counts()
location_counts.plot(kind='bar', xlabel='Location', ylabel='Frequency', title='Location Frequencies')
plt.show()
```

```python
#@title Location Frequencies by Detected Movement

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Security_Camera_Logs.csv')

df['Timestamp'] = pd.to_datetime(df['Timestamp'])

plt.figure(figsize=(10,6))
location_movement_counts = df.groupby(['Location', 'Detected Movement']).size().unstack()
location_movement_counts.plot(kind='bar', stacked=True, xlabel='Location', ylabel='Frequency', title='Location Frequencies by Detected Movement')
plt.show()
```
