```python
#@title Step 1: Importing Libraries
import pandas as pd #Data Manipulation
import numpy as np #Mathematical Calculation / Numerical Operation
import matplotlib.pyplot as plt #Plotting / Data Visualization
from sklearn.linear_model import LinearRegression #Model Making / Data Modeling
```

```python
#@title Step 2: Loading Data Set
df = pd.read_csv('Lesson_4_student_grades_dataset 2.csv')
df.head(10000)
```

```python
#@title Step 3: Visualizing the Data

xlabel = "Previous_Exam_Scores" #@param {type:"string"}
ylabel = "Grades" #@param {type:"string"}
title = "Hours of Study as (X)" #@param {type:"string"}
plt.scatter(df[xlabel], df[ylabel],color="green")
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.title(title)
plt.show()
```

```python
#@title Step 4: Preparing Data for Modeling
X = df[['Previous_Exam_Scores']]
Y = df[['Grades']]
```

```python
#@title Step 5: Fitting the Linear Regression Model
model = LinearRegression()
model.fit(X,Y)
```

```python
#@title Step 6: Getting Model Coefficient
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope: {slope}\nIntercept: {intercept}")
```

```python
#@title Step 7: Visualizing the Linear Progression Line
Show_Result = True #@param {type:"boolean"}
if Show_Result == True:
  plt.scatter(X,Y, color="green")
  plt.plot(X, model.predict(X), color="red")
  xlabel = "Independent Variable (X)" #@param {type:"string"}
  ylabel = "Dependent Variable (Y)" #@param {type:"string"}
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  plt.title('Scatter Plot of Data')
  plt.show()
```

```python
#@title Step 8:  Making Predictions
Range_From = 99 #@param {type:"integer"}
Range_To = 100 #@param {type:"integer"}
new_X = [[np.random.randint(Range_From, Range_To)]]
predicted_Y = model.predict(new_X)
print(f"\n\n\n\n\nValue of X: {new_X}\nPredicted Value of Y: {predicted_Y}\n\n\n\n\n\n\n")
```
